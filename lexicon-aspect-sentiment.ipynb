```python
def weighted(nb):
    if nb is None:
        return -float('inf')
    else:
        return nb
cat_value={}
```

```python
list1=[u'Battery', u'design', u'price', u'miscellaneous', u'quality',u'operation',u'service']
max_score = -100
max_sim = ''
cats_dict={}
for sid in aspects.keys():
    l=aspects[sid]
    #print(l)
    similarities=[]
    inner=[]
    categories=[]
    for asp in l:
        #print asp
        inner=[]
        sense1=wordnet.synsets(asp)
        #print "sense1:",sense1
        for cat in list1:
            #print cat
            sense2=wordnet.synsets(cat)
            #print "sense2",sense2
            for s1,s2 in product(sense1,sense2):
                score=wordnet.wup_similarity(s1,s2)
                #print(score)
                # print score,s1,s2
                if(score!=None):
                    inner.append((score,s2))
        #print "\n"
        if inner and len(inner)>0:
            
            topcat=sorted(inner,key=itemgetter(0),reverse=True)[0]
            categories.append(topcat[1].name()[:topcat[1].name().index('.')])
            cat_value[sid]=(topcat[0])
            #print(inner)
            #print(topcat)
    cats_dict[sid]=list(set(categories))
```

```python
def get_window(sentence, phrase, window_size):
    sentence = sentence.split()
    phrase = phrase.split()
    words = len(phrase)

    for i,word in enumerate(sentence):
        if word == phrase[0] and sentence[i:i+words] == phrase:
            start = max(0, i-window_size)
            return ' '.join(sentence[start:i+words+window_size])
```
